```python
'''
Q1. 
Explain following with an example : 
    
1 Artificial Intelligence
2. Machine Learning 
3. Deep Learning
'''

Artificial Intelligence : 
    
Artificial Intelligence is a broad field of computer science that involves the development 
of intelligent machines that can perform tasks that typically require human intelligence, 
such as speech recognition, language translation, and decision-making. AI systems can be 
trained using machine learning techniques to improve their performance over time. 
An example of AI would be a self-driving car that uses sensors, computer vision, and
machine learning algorithms to navigate roads and make decisions about when to brake, accelerate, or turn.


Machine Learning (ML):

Machine Learning is a subfield of AI that involves training machines to learn from data without being explicitly programmed. 
ML algorithms can automatically identify patterns and relationships in data and use that knowledge to make predictions or 
take actions. An example of ML would be a spam filter that learns to recognize spam emails by analyzing thousands of examples
of spam and non-spam emails.

Deep Learning (DL):

Deep Learning is a subfield of ML that involves training artificial neural networks to learn from data. DL algorithms can 
automatically learn to identify complex patterns and relationships in data, making them well-suited for tasks such as image
recognition, natural language processing, and speech recognition. An example of DL would be a computer vision system that can
recognize and classify images of cats and dogs based on millions of labeled examples of cats and dogs.
```

```python
'''
Q2. Explain supervised learning and give some examples . 
'''

Supervised Machine Learning:
    
Supervised learning is a type of machine learning where the algorithm is trained on labeled data. In other words, the algorithm 
is given input/output pairs and learns to map the inputs to the correct outputs. For example, you could train an algorithm to 
predict whether an email is spam or not by feeding it a dataset of emails labeled as "spam" or "not spam." The algorithm would 
learn to recognize patterns in the emails that are associated with spam, and use those patterns to make predictions on new, 
unlabeled emails.

Examples: 
Image classification: 
    Given a dataset of labeled images (e.g. pictures of dogs and cats), the algorithm 
    is trained to predict the correct label for new images.
    
Spam detection: 
    Given a dataset of labeled emails (e.g. spam and non-spam), the algorithm is 
    trained to classify new emails as either spam or not spam.
Stock price prediction: 
    Given a dataset of historical stock prices, the algorithm 
    is trained to predict the future price of a stock.
```

```python
'''
Q3. Explain Unsupervised learning and give some examples . 
'''
Unsupervised Machine Learning:
    
Unsupervised learning is a type of machine learning where the algorithm is trained on unlabeled data. In other words, 
the algorithm is given input data without any corresponding output labels and must find patterns and relationships within
the data on its own. For example, you could use unsupervised learning to cluster customers based on their buying habits without 
any prior knowledge of which customers belong to which clusters.

Clustering: 
    Given a dataset of unlabeled data points, the algorithm is trained to 
    group similar data points together into clusters.
Anomaly detection:
    Given a dataset of unlabeled data points, the algorithm is trained to 
    detect data points that are significantly different from the rest of the data.
Dimensionality reduction: 
    Given a dataset with a large number of features, the algorithm is trained to
    find a smaller set of features that capture the most important information in the data.
```

```python

'''
Q4. Explain semi supervised learning and give some examples . 
'''
Semi-Supervised Machine Learning:
    
Semi-supervised learning is a combination of supervised and unsupervised learning. The algorithm is trained on a dataset that
contains both labeled and unlabeled data. The labeled data is used to train the algorithm as in supervised learning, while the 
unlabeled data is used to find patterns and relationships as in unsupervised learning. For example, if you only have a small
dataset of labeled emails, you could use semi-supervised learning to train an algorithm to classify emails by also using a larger 
dataset of unlabeled emails to find patterns and relationships that could help improve the classification accuracy.

Speech recognition: 
    Given a dataset of labeled speech samples and a larger dataset of unlabeled speech samples,
    the algorithm is trained to recognize speech more accurately by incorporating the unlabeled data.
Text classification: 
    Given a small labeled dataset of text documents and a larger unlabeled dataset of text documents,
    the algorithm is trained to classify text more accurately by leveraging the additional unlabeled data.
Medical diagnosis: 
    Given a small labeled dataset of medical images and a larger unlabeled dataset of medical images,
    the algorithm is trained to improve the accuracy of medical diagnosis by incorporating the additional unlabeled data.
```

```python
'''
Q6 . What is Train , validation and Test dataset , explain it with some example 
'''


in machine learning, the dataset is typically split into three subsets: the training set, the validation set,
and the test set. Here an explanation of each subset:

Training set:
    The training set is the portion of the dataset used to train the machine learning model. It is the part of the 
    dataset on which the model learns the relationships and patterns between the input features and the output variable. 
    The training set is used to adjust the parameters of the model, so it can make accurate predictions on new, unseen data.

Validation set:
    The validation set is a subset of the dataset used to evaluate the performance of the model during training.
    It is used to tune the hyperparameters of the model and to prevent overfitting, which occurs when the model 
    fits the training data too well, but doesn generalize well to new data. The validation set allows us to estimate
    the performance of the model on new data and to make changes to the model if needed.

Test set:
    The test set is a subset of the dataset that is used to evaluate the performance of the model after it has been
    trained. It is used to get an unbiased estimate of the performance of the model on new, unseen data. The test
    set should be separate from the training and validation sets and should only be used once, after the model has been
    trained and fine-tuned.

    The typical split of the dataset is 70:15:15 or 80:10:10, where 70-80% of the data is used for training, 15-10% 
    is used for validation, and the remaining 15-10% is used for testing. It important to note that the split ratios 
    can vary depending on the size and complexity of the dataset, as well as the goals of the project.

In summary, the training set is used to train the model, the validation set is used to fine-tune the model 
```

```python
'''
Q7 How can supervised learning is used as anomaly detection . 
'''

Supervised learning can be used for anomaly detection by labeling anomalous instances 
as a separate class and training a classification model to distinguish between normal 
and anomalous instances. The model can then be used to detect anomalies in new data by 
predicting whether an instance belongs to the normal or anomalous class.
```
